```python
import pandas as pd
```

```python
regis_rev_2019 = pd.read_csv("data/2019_sum_regit.csv")
regis_rev_2022 = pd.read_csv("data/2022_sum_regit.csv")
```

```python
merged_df = pd.merge(regis_rev_2019,regis_rev_2022, on = 'dist_code',suffixes=('_2021', '_2022'))
```

```python
merged_df['percentage_growth'] = ((merged_df['sum(documents_registered_rev)_2022'] - merged_df['sum(documents_registered_rev)_2021']) / merged_df['sum(documents_registered_rev)_2021']) * 100
sorted_df = merged_df.sort_values(by='percentage_growth', ascending=False)
sorted_df.drop(columns= ['sum(estamps_challans_rev)_2021','sum(estamps_challans_rev)_2022'])
sorted_df.to_csv('regis_growth.csv', index = False)
```

```python
stamp_category = pd.read_csv("data/stamps_renv_category.csv")
```

```python
stamp_category
```

```python
#answering question 4

stamp_category['revenue'] = (stamp_category['sum(estamps_challans_rev)'] + stamp_category['sum(documents_registered_rev)']) 

# Calculate quartiles
quartiles = stamp_category['revenue'].quantile([0.25, 0.75])


# Define segment categories
low_segment = stamp_category[stamp_category['revenue'] < quartiles[0.25]]
medium_segment = stamp_category[(stamp_category['revenue'] >= quartiles[0.25]) & (stamp_category['revenue'] < quartiles[0.75])]
high_segment = stamp_category[stamp_category['revenue'] >= quartiles[0.75]]

# Visualize the distribution
import matplotlib.pyplot as plt
segment_counts = [len(low_segment), len(medium_segment), len(high_segment)]
segments = ['Low Revenue', 'Medium Revenue', 'High Revenue']

plt.bar(segments, segment_counts)
plt.xlabel('Revenue Segment')
plt.ylabel('Number of Districts')
plt.title('Distribution of Districts by Revenue Segment')
plt.show()
```
